```python
# plt.plot(x, y colot = 'color name' , linestyle = 'line_style' linewidth = value , marker = 'marker symbol', label = 'label name' )
# plt.xlabel

import matplotlib.pyplot as plt

months = [1, 2, 3, 4]
sales = [1000, 1500, 1200, 1800]

plt.plot(
    months,
    sales,
    color='green',
    linestyle=':', 
    linewidth=3,
    marker='o',
    label='2025 sales data'
)

plt.xlabel('Months')
plt.ylabel('Sales per Month')
plt.title(' Monthly Sales Data Report')
plt.legend()
plt.grid(color = 'black', linewidth = '1')
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()

```

# Bar Graph

```python
import matplotlib.pyplot as plt

product = ['A','B','C','D']
Sales = [1000,1000,800,1200]

plt.bar(product , Sales , color = 'orange' , label = 'Sales 2025')

plt.xlabel('product')
plt.ylabel('Sales')

plt.title('product Sale comparison')
plt.grid()

plt.legend()
plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt

regions = ['North', 'South' , 'East' , 'West']
revenue = [3000  , 2000 , 1600 , 1000]

plt.pie(revenue , labels=regions , autopct='%1.1f%%',colors=['gold' , 'skyblue','lightgreen','coral'])
plt.title('Revenue Contribution By Region')
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt

scores = [45,58,43,22,77,87,55,98,77,87,98,65,43,23,64,83,75,38,43,27]

plt.hist(scores, bins=5, color='purple', edgecolor='black')

plt.xlabel('Score Range')
plt.ylabel('No. of Student')
plt.title('Score Distribution of Student')

plt.show()
```

# Scatter

```python
#plt.scatter(x,y,color='color_name' , marker = 'marker style', label = 'label name')

hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]

plt.scatter(hours_studies, exam_scores , color='green' , marker='^',label='Student Data')

plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title("Relationship between study Time and Exam score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#plt.scatter(x,y,color='color_name' , marker = 'marker style', label = 'label name')

plt.scatter([1,2,3], [50,55,60] , color= 'black',label = 'Class A ')
plt.scatter([1,2,3], [45,50,52] , color= 'orange',label = 'Class B ')


plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title("Comparision of Two Classes")
plt.legend()
plt.grid(True)
plt.show()
```

# SubPlots

```python


x = [1,2,3,4]
y = [10,20,30,40]

plt.subplot(1,2,1)
plt.plot(x, y)
plt.title("Line Chart")

plt.subplot(1,2,2)
plt.bar(x, y)
plt.title("Bar Chart")

plt.show()
```

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots(1,2, figsize = (10,5))

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y,color='green')
ax[0].set_title('Line Plot')

ax[1].bar(x,y)
ax[1].set_title('Bar Chart')

fig.suptitle('Comparison of Line and Bar Charts')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,25]


plt.plot(x,y, color='blue',marker='o')
plt.title('Simple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.savefig('Line_plot.png',dpi = 300,bbox_inches = 'tight')
plt.show()
```
