```python
# plt.plot(x, y colot = 'color name' , linestyle = 'line_style' linewidth = value , marker = 'marker symbol', label = 'label name' )
# plt.xlabel

import matplotlib.pyplot as plt

months = [1, 2, 3, 4]
sales = [1000, 1500, 1200, 1800]

plt.plot(
    months,
    sales,
    color='green',
    linestyle=':', 
    linewidth=3,
    marker='o',
    label='2025 sales data'
)

plt.xlabel('Months')
plt.ylabel('Sales per Month')
plt.title(' Monthly Sales Data Report')
plt.legend()
plt.grid(color = 'black', linewidth = '1')
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()

```

```python
import matplotlib.pyplot as plt

product = ['A','B','C','D']
Sales = [1000,1000,800,1200]

plt.bar(product , Sales , color = 'orange' , label = 'Sales 2025')

plt.xlabel('product')
plt.ylabel('Sales')

plt.title('product Sale comparison')
plt.grid()

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

regions = ['North', 'South' , 'East' , 'West']
revenue = [3000  , 2000 , 1600 , 1000]

plt.pie(revenue , labels=regions , autopct='%1.1f%%',colors=['gold' , 'skyblue','lightgreen','coral'])
plt.title('Revenue Contribution By Region')
plt.show()
```
